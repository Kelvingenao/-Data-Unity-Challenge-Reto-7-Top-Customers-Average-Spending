```python
# üì¶ Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# üéØ Visual settings
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# üì• Load the dataset
df = pd.read_csv("supermarket_sales_categories.csv")

# üîç Initial exploration
print(df.head())
print(df.info())
print(df.describe())

# üßÆ Total spending per customer
spending_per_customer = df.groupby("CustomerID")["TotalSales"].sum().reset_index()
spending_per_customer.columns = ["CustomerID", "TotalSpending"]
spending_per_customer = spending_per_customer.sort_values(by="TotalSpending", ascending=False)

# üìä Top 5 customers
top_5 = spending_per_customer.head(5)
print("üèÜ Top 5 Customers:\n", top_5)

# üìà Bar chart ‚Äì Top 5
sns.barplot(data=top_5, x="CustomerID", y="TotalSpending", palette="viridis")
plt.title("Top 5 Customers by Total Spending")
plt.xlabel("CustomerID")
plt.ylabel("Total Spending")
plt.tight_layout()
plt.show()

# üìâ Histogram ‚Äì Spending distribution
sns.histplot(spending_per_customer["TotalSpending"], bins=30, kde=True, color="skyblue")
plt.title("Distribution of Total Spending per Customer")
plt.xlabel("Total Spending")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# üìä Average spending
average_spending = spending_per_customer["TotalSpending"].mean()
print(f"üí∞ Average spending per customer: ${average_spending:.2f}")

# üßæ Conclusion
print("\nüìå Conclusion:")
print("The top 5 customers represent a significant portion of total sales.")
print(f"The average spending per customer is ${average_spending:.2f}, which can guide segmentation and loyalty strategies.")
```
